# 2. Power Analysis (usually the 'Total' or 'Attack' column)

```python
# --- Power Analysis ---

# Calculate basic statistics for attack
print("\n--- Power Analysis ---")
stats = df_rock['attack'].describe()

# Convert statistics Series to DataFrame for tabulate
stats_df = pd.DataFrame(stats).reset_index()
stats_df.columns = ['Statistic', 'Value']

# Display table with tabulated format
print(tabulate(stats_df, headers='keys', tablefmt='fancy_grid', showindex=False))

# Create histogram for attack distribution
plt.figure(figsize=(10, 6))
sns.histplot(df_rock['attack'], kde=True, bins=10)
plt.title('Attack Distribution for Rock-type Pokémon')
plt.xlabel('Attack')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()

# Find the most powerful and weakest Rock-type Pokémon based on attack
most_powerful = df_rock.loc[df_rock['attack'].idxmax()]
weakest = df_rock.loc[df_rock['attack'].idxmin()]

# Create comparison table
attack_table = [
    ["Most Powerful", most_powerful["name"], most_powerful["attack"]],
    ["Weakest", weakest["name"], weakest["attack"]]
]

# Print result
print("\n--- Rock-type Pokémon Attack Comparison ---")
print(tabulate(attack_table, headers=["Type", "Name", "Attack"], tablefmt="fancy_grid"))

# Create bar chart of top 5 Rock-type Pokémon by attack
top_5_attack = df_rock.sort_values('attack', ascending=False).head(5)
plt.figure(figsize=(12, 6))
sns.barplot(x='name', y='attack', data=top_5_attack)
plt.title('Top 5 Rock-type Pokémon by Attack')
plt.xlabel('Pokémon')
plt.ylabel('Attack')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(True, alpha=0.3)
plt.show()
```

# 3. Defense Analysis

```python
# Calculate basic statistics for defense
print("\n--- Defense Analysis ---")
stats_def = df_rock['defense'].describe()

# Convert the Series to a DataFrame
stats_def_df = pd.DataFrame(stats_def).reset_index()
stats_def_df.columns = ['Statistic', 'Value']

# Pretty print using tabulate
print(tabulate(stats_def_df, headers='keys', tablefmt='fancy_grid', showindex=False))

# Create histogram for defense distribution
plt.figure(figsize=(10, 6))
sns.histplot(df_rock['defense'], kde=True, bins=10)
plt.title('Defense Distribution for Rock-type Pokémon')
plt.xlabel('Defense')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()

# Find the most defensive Rock-type Pokémon (highest defense)
most_defensive = df_rock.loc[df_rock['defense'].idxmax()]

# Find the least defensive Rock-type Pokémon (lowest defense)
least_defensive = df_rock.loc[df_rock['defense'].idxmin()]

# Create table to show the result
defense_table = [
    ["Most Defensive", most_defensive["name"], most_defensive["defense"]],
    ["Least Defensive", least_defensive["name"], least_defensive["defense"]]
]

# Print the table with tabulate
print("\n--- Rock-type Pokémon Defense Comparison ---")
print(tabulate(defense_table, headers=["Type", "Name", "Defense"], tablefmt="fancy_grid"))

print(f"\nMost defensive Rock Pokémon: {most_defensive['name']} with {most_defensive['defense']} defense")
print(f"Least defensive Rock Pokémon: {least_defensive['name']} with {least_defensive['defense']} defense")

# Create a bar chart of the top 5 rock Pokémon by defense
top_5_defense = df_rock.sort_values('defense', ascending=False).head(5)
plt.figure(figsize=(12, 6))
sns.barplot(x='name', y='defense', data=top_5_defense)
plt.title('Top 5 Rock-type Pokémon by Defense')
plt.xlabel('Pokémon')
plt.ylabel('Defense')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### IQR Method Data

```python
# Function to find outliers using IQR
def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Find outliers for speed, height, and weight
speed_outliers = find_outliers_iqr(df_rock, 'speed')
height_outliers = find_outliers_iqr(df_rock, 'height_m')
weight_outliers = find_outliers_iqr(df_rock, 'weight_kg')

# Display outlier analysis
print("\n--- Outlier Analysis ---")

# Speed outliers analysis
print(f"\nNumber of speed outliers: {len(speed_outliers)}")
if len(speed_outliers) > 0:
    print("Speed outliers:")
    speed_outliers_table = [(row['name'], row['speed']) for _, row in speed_outliers.iterrows()]
    print(tabulate(speed_outliers_table, headers=["Pokémon", "Speed"], tablefmt="fancy_grid"))

# Height outliers analysis
print(f"\nNumber of height outliers: {len(height_outliers)}")
if len(height_outliers) > 0:
    print("Height outliers:")
    height_outliers_table = [(row['name'], row['height_m']) for _, row in height_outliers.iterrows()]
    print(tabulate(height_outliers_table, headers=["Pokémon", "Height (m)"], tablefmt="fancy_grid"))

# Weight outliers analysis
print(f"\nNumber of weight outliers: {len(weight_outliers)}")
if len(weight_outliers) > 0:
    print("Weight outliers:")
    weight_outliers_table = [(row['name'], row['weight_kg']) for _, row in weight_outliers.iterrows()]
    print(tabulate(weight_outliers_table, headers=["Pokémon", "Weight (kg)"], tablefmt="fancy_grid"))
```

```python
# Create a comprehensive dataframe with key metrics
df_rock_stats = df_rock[['name', 'type1', 'type2', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'base_total', 'is_legendary']]

# Find fastest and slowest rock Pokémon
fastest = df_rock.loc[df_rock['speed'].idxmax()]
slowest = df_rock.loc[df_rock['speed'].idxmin()]

# Create a table with means and medians
fastest_and_slowest = [
    ["Fastest Pokémon", fastest['name'], f"{fastest['speed']}"],
    ["Slowest Pokémon", slowest['name'], f"{slowest['speed']}"]
]

# Print a statistics table
print(tabulate(fastest_and_slowest, headers=["Description", "Pokémon", "Speed"], tablefmt="fancy_grid"))

# Create a copy explicitly
df_rock_stats = df_rock[['name', 'type1', 'type2', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'base_total', 'is_legendary']].copy()

# Calculate a score based on key battle statistics
df_rock_stats['battle_score'] = df_rock_stats['hp'] * 0.1 + df_rock_stats['attack'] * 0.2 + df_rock_stats['defense'] * 0.2 + df_rock_stats['sp_attack'] * 0.15 + df_rock_stats['sp_defense'] * 0.15 + df_rock_stats['speed'] * 0.2

# Ordenar por battle_score en orden descendente
df_rock_stats = df_rock_stats.sort_values('battle_score', ascending=False)

# Mostrar los 10 mejores Pokémon Roca por Battle Score
print("\n--- Strategic Choice Analysis ---")
print("Top 10 Rock-type Pokémon by Battle Score:")
print(tabulate(df_rock_stats.head(10), headers='keys', tablefmt='fancy_grid', showindex=False))

# Create a bar chart for the top 10 Pokémon by battle score
plt.figure(figsize=(12, 6))
sns.barplot(x='name', y='battle_score', data=df_rock_stats.head(10))
plt.title('Top 10 Rock-type Pokémon by Battle Score')
plt.xlabel('Pokémon')
plt.ylabel('Battle Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(True, alpha=0.3)
plt.show()

# Create a radar chart for the top choice
top_choice = df_rock_stats.iloc[0]
print(f"\nTop strategic choice: {top_choice['name']}")
print(f"Type: {top_choice['type1']}/{top_choice['type2']}")
print(f"Stats: HP={top_choice['hp']}, Attack={top_choice['attack']}, Defense={top_choice['defense']}, Sp. Attack={top_choice['sp_attack']}, Sp. Defense={top_choice['sp_defense']}, Speed={top_choice['speed']}")
print(f"Legendary: {'Yes' if top_choice['is_legendary'] else 'No'}")

# Create radar chart for top 3 choices
top_3 = df_rock_stats.head(3)
categories = ['HP', 'Attack', 'Defense', 'Sp. Attack', 'Sp. Defense', 'Speed']
N = len(categories)

# Create angle for each category
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]  # Close the loop

# Create radar chart
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)

for i, pokemon in top_3.iterrows():
    values = [pokemon['hp'], pokemon['attack'], pokemon['defense'],
              pokemon['sp_attack'], pokemon['sp_defense'], pokemon['speed']]
    values += values[:1]  # Close the loop

    # Plot values
    ax.plot(angles, values, linewidth=2, label=pokemon['name'])
    ax.fill(angles, values, alpha=0.1)

# Set category labels
plt.xticks(angles[:-1], categories)

# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.title('Top 3 Rock-type Pokémon Comparison')
plt.show()

# Análisis de combate estratégico
print("\n--- Strategic Battle Analysis ---")
print("Based on the type effectiveness analysis:")

# Detalles de enfrentamientos favorables y desfavorables
favorable_matchup = ', '.join([key for key, value in rock_effectiveness.items() if value > 1])
unfavorable_matchup = ', '.join([key for key, value in rock_effectiveness.items() if value < 1])

# Mostrar detalles usando tabulate
battle_analysis_data = [
    ["Favorable matchup", favorable_matchup],
    ["Unfavorable matchup", unfavorable_matchup],
    ["Most vulnerable to", most_damage_type],
    ["Resist attacks from", least_damage_type],
]

# Imprimir análisis de batalla con tabulate
print(tabulate(battle_analysis_data, headers=["Analysis", "Details"], tablefmt="fancy_grid"))

# Resumen
print("\n--- Summary ---")
summary_data = [
    ["Number of Rock-type Pokémon", len(df_rock)],
    ["Legendary Rock-type Pokémon", len(legendary_rock)],
    ["Strongest attacker", f"{most_powerful['name']} ({most_powerful['attack']} attack)"],
    ["Best defender", f"{most_defensive['name']} ({most_defensive['defense']} defense)"],
    ["Fastest", f"{fastest['name']} ({fastest['speed']} speed)"],
    ["Top strategic choice", top_choice['name']],
]

# Imprimir resumen con tabulate
print(tabulate(summary_data, headers=["Metric", "Details"], tablefmt="fancy_grid"))
```
