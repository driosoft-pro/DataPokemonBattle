# 1. Basic Count

```python
# Filter Rock-type Pokémon
df_rock = df[(df['type1'] == 'rock') | (df['type2'] == 'rock')]
print(f"Total Rock-type Pokémon: {len(df_rock)}")
```

# 2. Power Analysis (usually the 'Total' or 'Attack' column)

```python
# Calculate basic statistics for attack
print("\n--- Power Analysis ---")
print("Attack statistics:")
print(df_rock['attack'].describe())

# Create histogram for attack distribution
plt.figure(figsize=(10, 6))
sns.histplot(df_rock['attack'], kde=True, bins=10)
plt.title('Attack Distribution for Rock-type Pokémon')
plt.xlabel('Attack')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()

# Create boxplot for attack
plt.figure(figsize=(10, 6))
sns.boxplot(y=df_rock['attack'])
plt.title('Attack Boxplot for Rock-type Pokémon')
plt.ylabel('Attack')
plt.grid(True, alpha=0.3)
plt.show()

# Find the most powerful and weakest rock Pokémon based on attack
most_powerful = df_rock.loc[df_rock['attack'].idxmax()]
weakest = df_rock.loc[df_rock['attack'].idxmin()]

print(f"\nMost powerful Rock Pokémon (by attack): {most_powerful['name']} with {most_powerful['attack']} attack")
print(f"Weakest Rock Pokémon (by attack): {weakest['name']} with {weakest['attack']} attack")

# Create a bar chart of the top 10 rock Pokémon by attack
top_10_attack = df_rock.sort_values('attack', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='name', y='attack', data=top_10_attack)
plt.title('Top 10 Rock-type Pokémon by Attack')
plt.xlabel('Pokémon')
plt.ylabel('Attack')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 3. Defense Analysis

```python
# Calculate basic statistics for defense
print("\n--- Defense Analysis ---")
print("Defense statistics:")
print(df_rock['defense'].describe())

# Create histogram for defense distribution
plt.figure(figsize=(10, 6))
sns.histplot(df_rock['defense'], kde=True, bins=10)
plt.title('Defense Distribution for Rock-type Pokémon')
plt.xlabel('Defense')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()

# Create boxplot for defense
plt.figure(figsize=(10, 6))
sns.boxplot(y=df_rock['defense'])
plt.title('Defense Boxplot for Rock-type Pokémon')
plt.ylabel('Defense')
plt.grid(True, alpha=0.3)
plt.show()

# Find the most defensive and least defensive rock Pokémon
most_defensive = df_rock.loc[df_rock['defense'].idxmax()]
least_defensive = df_rock.loc[df_rock['defense'].idxmin()]

print(f"\nMost defensive Rock Pokémon: {most_defensive['name']} with {most_defensive['defense']} defense")
print(f"Least defensive Rock Pokémon: {least_defensive['name']} with {least_defensive['defense']} defense")

# Create a bar chart of the top 10 rock Pokémon by defense
top_10_defense = df_rock.sort_values('defense', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='name', y='defense', data=top_10_defense)
plt.title('Top 10 Rock-type Pokémon by Defense')
plt.xlabel('Pokémon')
plt.ylabel('Defense')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 4. Gender Distribution (if you have a 'Percentage Male' column)

# 5. Legendary Pokémon

```python
# Count legendary rock Pokémon
legendary_rock = df_rock[df_rock['is_legendary'] == True]
print("\n--- Legendary Analysis ---")
print(f"Number of legendary Rock-type Pokémon: {len(legendary_rock)}")
print(f"Percentage of Rock-type Pokémon that are legendary: {(len(legendary_rock) / len(df_rock)) * 100:.2f}%")

# Display legendary rock Pokémon
print("\nLegendary Rock-type Pokémon:")
print(legendary_rock[['name', 'type1', 'type2']].to_string(index=False))

# Create bar chart for legendary vs non-legendary
plt.figure(figsize=(8, 6))
counts = [len(df_rock) - len(legendary_rock), len(legendary_rock)]
plt.bar(['Non-Legendary', 'Legendary'], counts, color=['#AEC6CF', '#FFD700'])
plt.title('Legendary vs Non-Legendary Rock-type Pokémon')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()
```

# 6. Relationship to other types (what types are weak/strong against)

```python
# Calculate average damage received from each type
damage_columns = [col for col in df.columns if 'against_' in col]
avg_damage = df_rock[damage_columns].mean().sort_values(ascending=False)

print("\n--- Type Relationship Analysis ---")
print("Average damage received from each type:")
print(avg_damage)

# Create bar chart for damage received
plt.figure(figsize=(14, 8))
sns.barplot(x=avg_damage.index, y=avg_damage.values)
plt.title('Average Damage Received by Rock-type Pokémon from Other Types')
plt.xlabel('Type')
plt.ylabel('Average Damage Multiplier')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(True, alpha=0.3)
plt.show()

# Find types that deal most and least damage to rock
most_damage_type = avg_damage.idxmax().replace('against_', '')
least_damage_type = avg_damage.idxmin().replace('against_', '')

print(f"\nType that deals most damage to Rock-type Pokémon: {most_damage_type} (multiplier: {avg_damage.max():.2f})")
print(f"Type that deals least damage to Rock-type Pokémon: {least_damage_type} (multiplier: {avg_damage.min():.2f})")

# For rock type Pokemon damage against other types
# This requires type effectiveness data
# Simplified version based on standard type effectiveness
print("\nRock-type Pokémon effectiveness against other types (simplified):")
rock_effectiveness = {
    'fire': 2.0,      # Super effective
    'ice': 2.0,       # Super effective
    'flying': 2.0,    # Super effective
    'bug': 2.0,       # Super effective
    'steel': 0.5,     # Not very effective
    'fighting': 0.5,  # Not very effective
    'ground': 0.5     # Not very effective
}

# Convert to Series for easier visualization
rock_effectiveness_series = pd.Series(rock_effectiveness)
print(rock_effectiveness_series)

# Create bar chart for rock effectiveness against other types
plt.figure(figsize=(10, 6))
sns.barplot(x=rock_effectiveness_series.index, y=rock_effectiveness_series.values)
plt.title('Rock-type Effectiveness Against Other Types')
plt.xlabel('Type')
plt.ylabel('Damage Multiplier')
plt.axhline(y=1, color='r', linestyle='-', alpha=0.3)  # Line at normal damage
plt.grid(True, alpha=0.3)
plt.show()
```

# 7. Speed, Height, Weight

```python
# Speed Analysis
print("\n--- Speed Analysis ---")
print("Speed statistics:")
print(df_rock['speed'].describe())

# Create histogram for speed
plt.figure(figsize=(10, 6))
sns.histplot(df_rock['speed'], kde=True, bins=10)
plt.title('Speed Distribution for Rock-type Pokémon')
plt.xlabel('Speed')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()

# Create boxplot for speed
plt.figure(figsize=(10, 6))
sns.boxplot(y=df_rock['speed'])
plt.title('Speed Boxplot for Rock-type Pokémon')
plt.ylabel('Speed')
plt.grid(True, alpha=0.3)
plt.show()

# Find fastest and slowest rock Pokémon
fastest = df_rock.loc[df_rock['speed'].idxmax()]
slowest = df_rock.loc[df_rock['speed'].idxmin()]

print(f"\nFastest Rock Pokémon: {fastest['name']} with {fastest['speed']} speed")
print(f"Slowest Rock Pokémon: {slowest['name']} with {slowest['speed']} speed")

# Height Analysis
print("\n--- Height Analysis ---")
print("Height statistics (in meters):")
print(df_rock['height_m'].describe())

# Create histogram for height
plt.figure(figsize=(10, 6))
sns.histplot(df_rock['height_m'].dropna(), kde=True, bins=10)
plt.title('Height Distribution for Rock-type Pokémon')
plt.xlabel('Height (m)')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()

# Create boxplot for height
plt.figure(figsize=(10, 6))
sns.boxplot(y=df_rock['height_m'].dropna())
plt.title('Height Boxplot for Rock-type Pokémon')
plt.ylabel('Height (m)')
plt.grid(True, alpha=0.3)
plt.show()

# Find tallest and shortest rock Pokémon
tallest = df_rock.loc[df_rock['height_m'].idxmax()]
shortest = df_rock.loc[df_rock['height_m'].idxmin()]

print(f"\nTallest Rock Pokémon: {tallest['name']} with {tallest['height_m']} meters")
print(f"Shortest Rock Pokémon: {shortest['name']} with {shortest['height_m']} meters")

# Weight Analysis
print("\n--- Weight Analysis ---")
print("Weight statistics (in kg):")
print(df_rock['weight_kg'].describe())

# Create histogram for weight
plt.figure(figsize=(10, 6))
sns.histplot(df_rock['weight_kg'].dropna(), kde=True, bins=10)
plt.title('Weight Distribution for Rock-type Pokémon')
plt.xlabel('Weight (kg)')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.show()

# Create boxplot for weight
plt.figure(figsize=(10, 6))
sns.boxplot(y=df_rock['weight_kg'].dropna())
plt.title('Weight Boxplot for Rock-type Pokémon')
plt.ylabel('Weight (kg)')
plt.grid(True, alpha=0.3)
plt.show()

# Find heaviest and lightest rock Pokémon
heaviest = df_rock.loc[df_rock['weight_kg'].idxmax()]
lightest = df_rock.loc[df_rock['weight_kg'].idxmin()]

print(f"\nHeaviest Rock Pokémon: {heaviest['name']} with {heaviest['weight_kg']} kg")
print(f"Lightest Rock Pokémon: {lightest['name']} with {lightest['weight_kg']} kg")

# Check outliers using IQR method
def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[(data[column] < (Q1 - 1.5 * IQR)) | (data[column] > (Q3 + 1.5 * IQR))]
    return outliers

# Find outliers for speed, height, and weight
speed_outliers = find_outliers_iqr(df_rock, 'speed')
height_outliers = find_outliers_iqr(df_rock, 'height_m')
weight_outliers = find_outliers_iqr(df_rock, 'weight_kg')

print("\n--- Outlier Analysis ---")
print(f"Number of speed outliers: {len(speed_outliers)}")
if len(speed_outliers) > 0:
    print("Speed outliers:")
    print(speed_outliers[['name', 'speed']].to_string(index=False))

print(f"\nNumber of height outliers: {len(height_outliers)}")
if len(height_outliers) > 0:
    print("Height outliers:")
    print(height_outliers[['name', 'height_m']].to_string(index=False))

print(f"\nNumber of weight outliers: {len(weight_outliers)}")
if len(weight_outliers) > 0:
    print("Weight outliers:")
    print(weight_outliers[['name', 'weight_kg']].to_string(index=False))

```

# 8. Strategic Choice

```python
# Create a comprehensive dataframe with key metrics
df_rock_stats = df_rock[['name', 'type1', 'type2', 'hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'base_total', 'is_legendary']]

# Find fastest and slowest rock Pokémon
fastest = df_rock.loc[df_rock['speed'].idxmax()]
slowest = df_rock.loc[df_rock['speed'].idxmin()]

print(f"\nFastest Rock Pokémon: {fastest['name']} with {fastest['speed']} speed")
print(f"Slowest Rock Pokémon: {slowest['name']} with {slowest['speed']} speed")

# Calculate a score based on key battle statistics
df_rock_stats['battle_score'] = df_rock_stats['hp'] * 0.1 + df_rock_stats['attack'] * 0.2 + df_rock_stats['defense'] * 0.2 + df_rock_stats['sp_attack'] * 0.15 + df_rock_stats['sp_defense'] * 0.15 + df_rock_stats['speed'] * 0.2

# Sort by battle score
df_rock_stats = df_rock_stats.sort_values('battle_score', ascending=False)

print("\n--- Strategic Choice Analysis ---")
print("Top 10 Rock-type Pokémon by Battle Score:")
print(tabulate.tabulate(df_rock_stats.head(10), headers='keys', tablefmt='pretty', showindex=False))

# Create a bar chart for the top 10 Pokémon by battle score
plt.figure(figsize=(12, 6))
sns.barplot(x='name', y='battle_score', data=df_rock_stats.head(10))
plt.title('Top 10 Rock-type Pokémon by Battle Score')
plt.xlabel('Pokémon')
plt.ylabel('Battle Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(True, alpha=0.3)
plt.show()

# Create a radar chart for the top choice
top_choice = df_rock_stats.iloc[0]
print(f"\nTop strategic choice: {top_choice['name']}")
print(f"Type: {top_choice['type1']}/{top_choice['type2']}")
print(f"Stats: HP={top_choice['hp']}, Attack={top_choice['attack']}, Defense={top_choice['defense']}, Sp. Attack={top_choice['sp_attack']}, Sp. Defense={top_choice['sp_defense']}, Speed={top_choice['speed']}")
print(f"Legendary: {'Yes' if top_choice['is_legendary'] else 'No'}")

# Create radar chart for top 3 choices
top_3 = df_rock_stats.head(3)
categories = ['HP', 'Attack', 'Defense', 'Sp. Attack', 'Sp. Defense', 'Speed']
N = len(categories)

# Create angle for each category
angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]  # Close the loop

# Create radar chart
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)

for i, pokemon in top_3.iterrows():
    values = [pokemon['hp'], pokemon['attack'], pokemon['defense'], 
              pokemon['sp_attack'], pokemon['sp_defense'], pokemon['speed']]
    values += values[:1]  # Close the loop
    
    # Plot values
    ax.plot(angles, values, linewidth=2, label=pokemon['name'])
    ax.fill(angles, values, alpha=0.1)

# Set category labels
plt.xticks(angles[:-1], categories)

# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.title('Top 3 Rock-type Pokémon Comparison')
plt.show()

print("\n--- Strategic Battle Analysis ---")
print("Based on the type effectiveness analysis:")
print(f"- Favorable matchup: Rock-type Pokémon perform well against {', '.join([key for key, value in rock_effectiveness.items() if value > 1])} types.")
print(f"- Unfavorable matchup: Rock-type Pokémon struggle against {', '.join([key for key, value in rock_effectiveness.items() if value < 1])} types.")
print(f"- Rock-type Pokémon are most vulnerable to {most_damage_type} attacks.")
print(f"- Rock-type Pokémon resist {least_damage_type} attacks.")

# Summary
print("\n--- Summary ---")
print(f"Number of Rock-type Pokémon: {len(df_rock)}")
print(f"Legendary Rock-type Pokémon: {len(legendary_rock)}")
print(f"Strongest attacker: {most_powerful['name']} ({most_powerful['attack']} attack)")
print(f"Best defender: {most_defensive['name']} ({most_defensive['defense']} defense)")
print(f"Fastest: {fastest['name']} ({fastest['speed']} speed)")
print(f"Top strategic choice: {top_choice['name']}")
```
